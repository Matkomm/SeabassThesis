```python
import pandas as pd
import plotly.graph_objects as go
#50% free will, 150 fish, 3600 time steps (1 hour)
# Load the new data format
fish_data_new = pd.read_csv(r'C:\Users\Mathiako\OneDrive - NTNU\Documents\Master Mathias\Master\Simulation_data\morning_final3.csv')
# Define the updated function
def create_2d_histogram(cluster_df, time_column: str = "time_step", depth_column: str = "pos_z", reverse_axis: bool = True):
    # Convert time steps to hours and minutes
    start_time = pd.Timestamp('2023-01-01 06:00:00')  # Starting at 06:00
    cluster_df['time_converted'] = start_time + pd.to_timedelta(cluster_df[time_column], unit='s')
    
    fig = go.Figure()
    fig.add_trace(go.Histogram2dContour(
        x=cluster_df['time_converted'],
        y=cluster_df[depth_column],
        colorscale='Blues',
        reversescale=True,
        nbinsx=64,
        nbinsy=48,
        xaxis='x',
        yaxis='y',
        colorbar={"title": 'Counts'}
    ))
    fig.add_hline(y=2, line_dash='dash', line_color='Red', line_width=4)
    if reverse_axis:
        fig.update_yaxes(autorange="reversed")
    fig.update_layout(
        xaxis_title="Time",
        yaxis_title="Depth in m",
        showlegend=True,
        autosize=False, # Disable autosizing
        width=1200, # Set the width of the figure
        height=700 # Set the height of the figure to make it more square
    )
    fig.update_xaxes(tickformat='%H:%M', dtick=300000*2)  # Set x-axis to show time in HH:MM:SS format
    fig.update_annotations(font_size=30)
    fig.update_layout(font_size=30)
    fig.show()

# Call the function with your data
create_2d_histogram(fish_data_new)
```

```python
import pandas as pd
import plotly.graph_objects as go
from plotly.subplots import make_subplots
#150 fish, 3600 time steps (1 hour), morning, noon, afternoon, evening
# Load the new data format
fish_data_morning = pd.read_csv(r'C:\Users\Mathiako\OneDrive - NTNU\Documents\Master Mathias\Master\Simulation_data\morning_final3.csv')
fish_data_noon = pd.read_csv(r'C:\Users\Mathiako\OneDrive - NTNU\Documents\Master Mathias\Master\Simulation_data\noon_final3.csv')
fish_data_afternoon = pd.read_csv(r'C:\Users\Mathiako\OneDrive - NTNU\Documents\Master Mathias\Master\Simulation_data\afternoon_final3.csv')
fish_data_evening = pd.read_csv(r'C:\Users\Mathiako\OneDrive - NTNU\Documents\Master Mathias\Master\Simulation_data\evening_final3.csv')

# Define the function to create a combined plot with separated time periods using subplots
def create_subplots_2d_histogram(morning_df, noon_df, afternoon_df, evening_df,
                                 morning_start_time: str = '2023-01-01 06:00:00',
                                 noon_start_time: str = '2023-01-01 10:00:00',
                                 afternoon_start_time: str = '2023-01-01 15:00:00',
                                 evening_start_time: str = '2023-01-01 19:00:00',
                                 reverse_axis: bool = True):
    # Convert time steps to hours and minutes for each period
    morning_start_time = pd.Timestamp(morning_start_time)
    morning_df['time_converted'] = morning_start_time + pd.to_timedelta(morning_df['time_step'], unit='s')
    
    noon_start_time = pd.Timestamp(noon_start_time)
    noon_df['time_converted'] = noon_start_time + pd.to_timedelta(noon_df['time_step'], unit='s')
    
    afternoon_start_time = pd.Timestamp(afternoon_start_time)
    afternoon_df['time_converted'] = afternoon_start_time + pd.to_timedelta(afternoon_df['time_step'], unit='s')
    
    evening_start_time = pd.Timestamp(evening_start_time)
    evening_df['time_converted'] = evening_start_time + pd.to_timedelta(evening_df['time_step'], unit='s')
    
    # Create subplots
    fig = make_subplots(rows=1, cols=4, shared_yaxes=True, subplot_titles=("Morning", "Noon", "Afternoon", "Evening"), horizontal_spacing=0.03)
    
    # Add morning data
    fig.add_trace(go.Histogram2dContour(
        x=morning_df['time_converted'],
        y=morning_df['pos_z'],
        colorscale='Blues',
        reversescale=True,
        nbinsx=64,
        nbinsy=48,
        zmax=900,
        zmin=0,
        showscale=False
    ), row=1, col=1)
    
    # Add noon data
    fig.add_trace(go.Histogram2dContour(
        x=noon_df['time_converted'],
        y=noon_df['pos_z'],
        colorscale='Blues',
        reversescale=True,
        nbinsx=64,
        nbinsy=48,
        zmax=900,
        zmin=0,
        showscale=False
    ), row=1, col=2)
    
    # Add afternoon data
    fig.add_trace(go.Histogram2dContour(
        x=afternoon_df['time_converted'],
        y=afternoon_df['pos_z'],
        colorscale='Blues',
        reversescale=True,
        nbinsx=64,
        nbinsy=48,
        zmax=900,
        zmin=0,
        showscale=False
    ), row=1, col=3)
    
    # Add evening data
    fig.add_trace(go.Histogram2dContour(
        x=evening_df['time_converted'],
        y=evening_df['pos_z'],
        colorscale='Blues',
        reversescale=True,
        nbinsx=64,
        nbinsy=48,
        colorbar={"title": 'Counts'},
        zmax=900,
        zmin=0
    ), row=1, col=4)
    
    # Add horizontal line for each plot
    for i in range(1, 5):
        fig.add_hline(y=2, line_dash='dash', line_color='Red', line_width=4, row=1, col=i)
    
    if reverse_axis:
        fig.update_yaxes(autorange="reversed")
    
    fig.update_layout(
        xaxis_title="Time",
        yaxis_title="Depth in m",
        showlegend=False,
        autosize=False, # Disable autosizing
        width=1800, # Set the width of the figure
        height=700 # Set the height of the figure to make it more square
    )
    fig.update_xaxes(tickformat='%H', dtick=3600000)
    fig.update_annotations(font_size=30)
    fig.update_layout(font_size=30)
    fig.show()

# Call the function with the separated morning, noon, afternoon, and evening data
create_subplots_2d_histogram(fish_data_morning, fish_data_noon, fish_data_afternoon, fish_data_evening)
```
